```python

# Step 2: Split features (x) and target (y)
x = df.drop(columns=['quality'])  # Replace 'quality' with your actual target column
y = df['quality']

# Step 3: Convert categorical variables in 'x' using one-hot encoding (if any)
x = pd.get_dummies(x, drop_first=True)

# Step 4: Convert both x and y to NumPy arrays and then to PyTorch tensors
x_tensor = torch.tensor(x.values, dtype=torch.float32)  # Convert x to a float32 tensor
y_tensor = torch.tensor(y.values, dtype=torch.float32)  # Convert y to a float32 tensor (ensure y is float if it's regression)

# Print to check if everything is converted properly
print("X Tensor Shape:", x_tensor.shape)
print("Y Tensor Shape:", y_tensor.shape)
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset

# Step 1: Create a TensorDataset
dataset = TensorDataset(x_tensor, y_tensor)

# Step 2: Create a DataLoader
data_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Step 3: Iterate through one batch of data
for batch_x, batch_y in data_loader:
    print("Batch X (features):", batch_x)
    print("Batch Y (targets):", batch_y)
    break  # Break after first batch to only print the first batch
```
